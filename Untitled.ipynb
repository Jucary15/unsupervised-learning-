```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path, index_col=0)
df.head()
```

```python

df.shape
```

```python
df[df['IsTrading']==True]
```

```python
df = df[df['IsTrading']==True]
df.head()
```

```python

df.info()
```

```python
df.isnull().sum()
```

```python
crypto_df = df.dropna()
crypto_df.head()
```

```python
crypto_df = crypto_df[crypto_df['TotalCoinsMined']>0]
```

```python
crypto_df
```

```python
crypto_df.drop(['IsTrading','CoinName'], axis=1, inplace=True)
crypto_df
```

```python
for i in crypto_df.columns:
    print(f'{i}: {crypto_df[i].nunique()}')
```

```python
X = pd.get_dummies(crypto_df, columns=['Algorithm', 'ProofType'])
X
```

```python
# Standarize data with StandarScaler
X_scaled = StandardScaler().fit_transform(X)
X_scaled
```

```python
X_scaled.shape
```

```python
# PCA to reduce dimensions and preserve 90% of the data
from sklearn.decomposition import PCA
# PCA model
pca = PCA(n_components=0.90)
# Two principal components for the iris data.
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
# t-SNE model on the dataframe X_pca to reduce dimensions to 2
from sklearn.manifold import TSNE
tsne = TSNE(perplexity=50)
```

```python
# Reduce dimensions
X_pca_tsne_features = tsne.fit_transform(X_pca)
X_pca_tsne_features

X_pca_tsne_features.shape
```

```python
plt.scatter(X_pca_tsne_features[:, 0], X_pca_tsne_features[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
k = list(range(1, 11))
# Inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(X_pca_tsne_features)
#     km.predict
    inertia.append(km.inertia_)
# Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
X_pca_tsne_features_elbow = pd.DataFrame(elbow_data)
X_pca_tsne_features_elbow.head(10)
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(X_pca_tsne_features_elbow['k'], X_pca_tsne_features_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('clusters')
plt.ylabel('inertia')
plt.title('curve for customer data')
plt.show()
```
